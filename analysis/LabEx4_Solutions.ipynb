### <span style = "color:blue">Question 2 (14 marks)</span>

```python
%reset

def simcal(num1, num2, ops):
    if (ops == "+"):
        result = num1 + num2
        return result
    elif(ops == "-"):
        result = num1 - num2
        return result
    elif(ops == "*"):
        result = num1 * num2
        return result
    else:
        result = num1 / num2
        return result

num1 = eval(input("Enter first number: "))
num2 = eval(input("Enter second number: "))
ops = input("Enter operation +,-,* or / to perform: ")

result = simcal(num1,num2,ops)
print("{} {} {} = {}".format(num1,ops,num2,result))
```
